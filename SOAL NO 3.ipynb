### Kondisi a


```python
query = [35, 1, 0, 1]
```

```python
#Perhitungan Distance Kondisi a
import math
dis = []
for i in range (80):
    dis.append(math.sqrt((float(df.iloc[i]['Usia']) - (float(query [0]))) **2 + 
                         (float(df.iloc[i]['Kelahiran_ke-']) - (float(query[1]))) **2 + 
                         (float(df.iloc[i]['Waktu_Kelahiran']) - (float(query[2]))) **2 + 
                         (float(df.iloc[i]['Tekanan_darah']) - (float(query[3]))) **2)
              )
```

```python
df['dis'] = dis
df
```

```python
df.sort_values('dis')
```

```python
#Yang terdekat ada di nomor 10, 11, 42, 53, 65
#Hitung mean
import statistics

caesarian = [1, 0, 0, 1, 1]
x = statistics.mean(caesarian)
print(x)
```

### Result a
